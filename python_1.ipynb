```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('/content/amazon.csv')
```

```python
# Check for and remove null values
print("Null values before removal:\n", data.isnull().sum())
data.dropna(inplace=True)
print("Null values after removal:\n", data.isnull().sum())
```

```python
data.corr()
```

```python
# Histogram of rating
plt.figure(figsize=(8, 6))
sns.histplot(data['rating'], bins=10, kde=True)
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter plot of discount_percentage vs rating
plt.figure(figsize=(8, 6))
sns.scatterplot(x='discount_percentage', y='rating', data=data)
plt.title('Discount Percentage vs Rating')
plt.xlabel('Discount Percentage')
plt.ylabel('Rating')
plt.show()
```
